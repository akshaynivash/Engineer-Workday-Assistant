Task 4:
Write a function that takes as input two timestamps of the form 2017/05/13 12:00 and 
calculates their differences in hours. Please only return the full hour difference and 
round the results. E.g. 2022/02/15 00:05 and 2022/02/15 01:00 would return 1 hour.

```python
from datetime import datetime

def hour_difference(timestamp1: str, timestamp2: str) -> int:
    # Define the format of the timestamps
    fmt = "%Y/%m/%d %H:%M"
    
    # Convert string timestamps to datetime objects
    dt1 = datetime.strptime(timestamp1, fmt)
    dt2 = datetime.strptime(timestamp2, fmt)
    
    # Calculate the absolute difference in seconds
    diff_seconds = abs((dt2 - dt1).total_seconds())
    
    # Convert to hours and round the result
    diff_hours = round(diff_seconds / 3600)
    
    return diff_hours

# Example usage:
ts1 = input("Enter first timestamp (YYYY/MM/DD HH:MM): ")
ts2 = input("Enter second timestamp (YYYY/MM/DD HH:MM): ")

print(hour_difference(ts1, ts2), "hours")
```

Task-5 
Expand the above function to only count the time difference between 09:00 – 17:00 and 
only on weekdays.


```python
from datetime import datetime, timedelta

def hour_difference(timestamp1: str, timestamp2: str) -> int:
    # Define the format of the timestamps
    fmt = "%Y/%m/%d %H:%M"
    
    # Convert string timestamps to datetime objects
    dt1 = datetime.strptime(timestamp1, fmt)
    dt2 = datetime.strptime(timestamp2, fmt)
    
    # Ensure dt1 is earlier than dt2
    if dt1 > dt2:
        dt1, dt2 = dt2, dt1
    
    # Define working hours
    work_start = 9
    work_end = 17
    
    total_work_hours = 0
    current = dt1
    
    while current < dt2:
        if current.weekday() < 5:  # Monday to Friday
            work_day_start = current.replace(hour=work_start, minute=0)
            work_day_end = current.replace(hour=work_end, minute=0)
            
            if current < work_day_start:
                current = work_day_start
            
            if current < work_day_end:
                end_time = min(dt2, work_day_end)
                total_work_hours += (end_time - current).total_seconds() / 3600
            
        current += timedelta(days=1)
        current = current.replace(hour=0, minute=0)
    
    return round(total_work_hours)

# Example usage:
ts1 = input("Enter first timestamp (YYYY/MM/DD HH:MM): ")
ts2 = input("Enter second timestamp (YYYY/MM/DD HH:MM): ")

result = hour_difference(ts1, ts2)
print(f"You entered: {ts1} and {ts2}")
print(f"Calculated working hours difference: {result} hours")
```
