```python
import pandas as pd

# Use the full absolute path
df = pd.read_csv(r"D:\bmw_intern\data\Parts.csv", delimiter=";")


# Display first few rows
print(df.head())

# Show column names
print(df.columns)
```

```python
# Check missing values
print(df.isnull().sum())


```

```python
# Check for duplicate descriptions
print("Duplicate Descriptions:", df.duplicated(subset=["DESCRIPTION"]).sum())
```

```python
import pandas as pd



# Remove rows where more than 50% of values are missing
threshold = 0.5 * len(df.columns)  # Calculate 50% threshold
df_cleaned = df.dropna(thresh=threshold)  # Drop rows with too many missing values


```

```python
df_cleaned
```

```python
import matplotlib.pyplot as plt

# Calculate the length of each description
df_cleaned['Description Length'] = df_cleaned['DESCRIPTION'].apply(lambda x: len(str(x)) if pd.notnull(x) else 0)

# Plot the description lengths
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['Description Length'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Description Lengths')
plt.xlabel('Description Length')
plt.ylabel('Frequency')
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all descriptions into a single string
text = ' '.join(df_cleaned['DESCRIPTION'].dropna().astype(str))

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plot the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Common Words in Descriptions')
plt.show()
```

```python
import seaborn as sns

# Select key numeric features
key_numeric_features = ['Description Length']  # Add other key numeric features if needed

# Plot the distribution of key numeric features
plt.figure(figsize=(12, 8))
for feature in key_numeric_features:
    sns.histplot(df_cleaned[feature], kde=True, bins=30, label=feature, alpha=0.6)
plt.title('Distribution of Key Numeric Features')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
